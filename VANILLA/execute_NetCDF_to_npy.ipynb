```python
import os
from os import path as op
from subprocess import Popen
import re
import NetCDF_to_npy as pkg

DATAPATH=''
```

```python
inPath = os.path.expanduser('/scratch/adomakor412/fall_2018')
outPath = os.path.expanduser('~/storage')
```

```python
def download(filename,toPath):
    cmd = f'python3 {pkg} {filename} {toPath}'
    bash = cmd.format({'pkg':{pkg}, 'filename':{filename},'toPath':{toPath}})
    print('bash: ',bash)
    pid = Popen(cmd, shell=True)
    pid.communicate()
    return
```

```python
cmd = f'cd {inPath}'
Popen(cmd.format({'inPath':inPath}), shell=True).communicate()

dirs = os.listdir( inPath )

newPath = outPath 

# This would print all the files and directories
for fileList in dirs:
    filePath = op.join(inPath, fileList)
    if os.path.isdir(filePath):
        # skip directories
        continue
    print(filePath)
    download(filePath,newPath)
```
