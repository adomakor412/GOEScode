```python
def listurls(prefix,html):
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html.text)
    urllist = [elt['href'] for elt in soup.find_all(href=re.compile(prefix))]
    return urllist
```

```python
for i in range(8,19):#the website for particular day
    dd = str(i).zfill(2) #all in the same month
    mystr = 'http://home.chpc.utah.edu/~u0553130/Brian_Blaylock/cgi-bin/goes16_download.cgi?source=aws&satellite=noaa-goes16&domain=F&product=ABI-L1b-Rad&date=2019-{mm}-{dd}&hour={hr}'
    mm = '04'
    for t in range(24):#this is for the hour
        for band in range(7,11):
            hr = str(t).zfill(2)
            mystr.format(**{'mm':mm, 'dd':dd, 'hr':hr})
            myurl = mystr + "t"
            req = requests.get(myURL)
            prefix = "OR_ABI-L1b-RadF-M6C" + str(band).zfill(2)
            listurls(prefix,req)
            f.write(str(listurls(prefix,req)))
            
            
```
